**Histogram: A histogram is a type of bar chart that represents the distribution of numerical data.
It groups data into bins (or intervals) and displays how many data points fall into each bin.
The height of each bar in the histogram corresponds to the frequency or count of data points within each bin.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data1 = pd.read_excel('News_release_chart_data_August_2024.xlsx')
```

```python
data1.head()
```

```python
plt.hist(data1['Apartment_unit'], bins=5, edgecolor='black')  

# Add labels and a title
plt.xlabel('House Price of Apartments Unit')
plt.ylabel('Frequency')
plt.title('Distribution of Apartment Units house prices')

# Display the histogram
plt.show()
```

Here are some observations:

X-axis (House Price of Apartment Unit): The range of prices spans from around 200,000 to 550,000. The prices are grouped into bins of approximately 50,000 increments, which means the data has been divided into price ranges like 200,000–250,000, 250,000–300,000, and so on.

Y-axis (Frequency): The height of the bars represents the frequency, or the number of apartment units that fall within each price range.

The highest bar corresponds to the price range 200,000–250,000, with around 70 units falling into this category.
The second highest range is 250,000–300,000, with slightly fewer units.
After 300,000, the frequency decreases significantly, with the smallest frequency in the range of 350,000–400,000.
There is a small increase again in the 450,000–500,000 and 500,000–550,000 ranges, showing a modest uptick in higher-priced units.
Distribution Shape:

The distribution is right-skewed, meaning the bulk of the apartment units are priced on the lower end of the spectrum (200,000–300,000), and fewer units are priced at the higher end (450,000 and above).
This indicates that most apartment units in this dataset are relatively affordable, but there is a small proportion of more expensive units.


This histogram suggests that the majority of apartments are concentrated in the lower price ranges, with a gradual tapering off in the number of units as prices increase. The data appears to represent a typical real estate distribution, where fewer units are available at higher price points.

Scatter Plot: A scatter plot is a type of data visualization used to display the values from two variables for a set of data. 
The data is displayed as a collection of points, each having the value of one variable determining the position on the 
horizontal axis and the value of the other variable determining the position on the vertical axis.

```python
plt.scatter(data1['Date'], data1['Apartment_unit'], color='blue', alpha=0.5, s=60, edgecolor='black')

# Add labels and a title
plt.xlabel('Apartment Units')
plt.ylabel('Housing Price')
plt.title('Scatter Plot of Apartment Units vs Price')

#showing the grid lines
plt.grid(True, linestyle='--', alpha=0.7)
# Display the scatter plot
plt.show()
```

The x-axis is labeled with years, running from 2008 to 2024.
The y-axis represents the housing price, ranging from 200,000 to 550,000.
Prices show a general upward trend over time, indicating an increase in apartment prices over the years.
This scatter plot represents the increasing cost of apartment units over time, with the largest price growth happening from 2012 to 2020. The consistent upward trend suggests a growing market, higher demand, or other economic factors pushing prices higher.
There is a positive correlation between time (as represented by the years) and housing prices. As time progresses, housing prices tend to rise, although there are fluctuations.
Between 2012 and 2016, the prices appear to be rising sharply, followed by more fluctuations after 2020.
Post 2020, there seems to be some variation and stabilization at higher price points, but the prices remain consistently high.
In summary, this scatter plot shows a time-based progression of increasing apartment unit prices, with fluctuations post-2020, likely reflecting market trends or changes in housing demand.

**Box-Whisker plot** : A box-and-whisker plot, or simply box plot, is a graphical representation of the distribution of a dataset that highlights its central tendency, variability, and potential outliers. It provides a summary of five important statistical measures:

Minimum: The smallest data point (excluding outliers).
First Quartile (Q1): The median of the lower half of the dataset (25th percentile).
Median (Q2): The middle value of the dataset (50th percentile).
Third Quartile (Q3): The median of the upper half of the dataset (75th percentile).
Maximum: The largest data point (excluding outliers).


```python
# Group by year, assuming there's a column named 'year' and 'price'
# Boxplot expects the data to be split by categories (in this case, 'year')
years = data1['Date']
prices = data1['Apartment_unit']

# Create a box plot
plt.figure(figsize=(10, 6))  # Adjust the size as needed
plt.boxplot([prices[years == year] for year in sorted(years.unique())], labels=sorted(years.unique()))

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Housing Price')
plt.title('Boxplot of Housing Prices by Year')

# Display the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if there are many years
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
